# Funções

```python
import sqlite3
from sqlite3 import Error

## Função de alteração (insert, update, delete...)
def execute_query(connection, query):
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        connection.commit()
        print(f"Query {query.split('(')[0].strip()} executada.")
        if cursor.rowcount != -1: print(f"{cursor.rowcount} linha(s) afetadas")

    except Error as e:
        print(f"Erro: '{e}'")


# Função de leitura de dados (select)
def execute_read_query(connection, query):
    cursor = connection.cursor()
    result = None
    try:
        cursor.execute(query)
        result = cursor.fetchall()

        return result
    except Error as e:
        print(f"Erro: '{e}'")
```

# Criando o DUMP

```python
## Criando conexão
connection = sqlite3.connect("gestaoFinanceira.sqlite")

query = """
CREATE TABLE Usuario(
	id INTEGER PRIMARY KEY AUTOINCREMENT NOT NULL,
    saldo decimal(65, 2),
    nome varchar(100) NOT NULL,
    sexo varchar(5),
    cpf char(20) unique NOT NULL,
    ativo tinyint(1) default 1,
    dataNascimento date
);
"""
execute_query(connection, query)

query = """
CREATE TABLE Categoria(
	id INTEGER PRIMARY KEY AUTOINCREMENT NOT NULL,
    nome varchar(100) NOT NULL,
    ativo tinyint(1) default 1
);
"""
execute_query(connection, query)


query = """
CREATE TABLE TransacaoEsporadica(
	id INTEGER PRIMARY KEY AUTOINCREMENT NOT NULL,
    nome varchar(100) NOT NULL,
    descricao varchar(500),
    dataDaTransacao date,
    ativo tinyint(1) default 1,
    valor decimal(65, 2),
    tipo varchar(10),
    categoria_id int NOT NULL,
    usuario_id int NOT NULL,

    CONSTRAINT fk_CategoriaTransacaoEsporadica FOREIGN KEY (categoria_id) REFERENCES Categoria(id),
    CONSTRAINT fk_UsuarioTransacaoEsporadica FOREIGN KEY (usuario_id) REFERENCES Usuario(id)
);
"""
execute_query(connection, query)

query = """
CREATE TABLE TransacaoRecorrente(
	id INTEGER PRIMARY KEY AUTOINCREMENT NOT NULL,
    nome varchar(100) NOT NULL,
    descricao varchar(500),
    diaVencimento int,
    ativo tinyint(1) default 1,
    valor decimal(65, 2),
    tipo varchar(10),
    categoria_id int NOT NULL,
    usuario_id int NOT NULL,

    CONSTRAINT fk_CategoriaTransacaoRecorrente FOREIGN KEY (categoria_id) REFERENCES Categoria(id),
    CONSTRAINT fk_UsuarioTransacaoRecorrente FOREIGN KEY (usuario_id) REFERENCES Usuario(id)
);
"""
execute_query(connection, query)

```

Crie um procedure relevante para sua aplicação em PI3. Descreva a ação esperada ao executar a procedure e, ao justificar o contexto, explique por que essa ação é tão recorrente no seu PI3 a ponto de precisar de uma procedure.

**-- EDITE ESSA CÉLULA --**

* **Justificativa para que a operação seja um Procedure (Contexto de PI3): Saber a quantidade de registros existentes de uma categoria em específico de forma recorrente para que o usuário possa visualizar.**
* **Descrição da ação esperada ao executar o procedure: Facilitar processos constantes de uma mesma consulta recorrente.**

```python
## ESCREVA ABAIXO O PROCEDURE
query = """
delimiter $$
CREATE PROCEDURE inserirTransacaoEsporadica(in nome varchar(100),
                                  in descricao varchar(500),
                                  in dataDaTransacao date,
								  in ativo tinyint(1),
                                  in valor decimal(65, 2),
                                  in tipo varchar(10),
                                  in categoria_id int,
								  in usuario_id int
								  )
BEGIN
	insert into
    transacaoEsporadica
            (nome, descricao, dataDaTransacao, ativo, valor, tipo, categoria_id, usuario_id)
    VALUES (nome, descricao, dataDaTransacao, ativo, valor, tipo, categoria_id, usuario_id);

	if tipo = 'gasto' then
		UPDATE usuario as u
		SET saldo = saldo - valor
		WHERE u.id = usuario_id;
	else
		UPDATE usuario as u
		SET saldo = saldo + valor
		WHERE u.id = usuario_id;
    end if;
END $$
delimiter ;
"""

execute_query(connection, query)
## ESCREVA ABAIXO UMA EXECUÇÃO DA PROCEDURE

query = """
CALL inserirTransacaoEsporadica(
    'Bônus de desempenho',
    'Premiação por metas',
    '2026-06-25',
    1,
    200.00,
    'ganho',
    10,
    2
);
"""

execute_query(connection, query)
```
